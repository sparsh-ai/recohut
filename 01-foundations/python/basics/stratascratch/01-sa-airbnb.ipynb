### Number Of Units Per Nationality

Find the number of apartments per nationality that are owned by people under 30 years old.
Output the nationality along with the number of apartments.
Sort records by the apartments count in descending order.

```python
airbnb_units = read_csv('airbnb_units')
airbnb_units.head()
```

```python
airbnb_hosts = read_csv('airbnb_hosts')
airbnb_hosts.head()
```

```python
merged = pd.merge(airbnb_units, airbnb_hosts, on='host_id')
merged = merged[(merged['age'] < 30) & (merged['unit_type']=='Apartment')]
result = merged.groupby(['nationality'])['unit_id'].nunique().to_frame(
'apartment_count').reset_index().sort_values('apartment_count')
result
```

### Find matching guests

Find matching hosts and guests pairs in a way that they are both of the same gender and nationality.
Output the host id and the guest id of matched pair.

```python
airbnb_guests = read_csv('airbnb_guests')
airbnb_guests.head()
```

```python
merged_df = pd.merge(airbnb_hosts,
                     airbnb_guests,
                     how='left',
                     left_on=['nationality','gender'],
                     right_on = ['nationality','gender'])
result = merged_df[['host_id','guest_id']].drop_duplicates()
result.head()
```

### Number Of Bathrooms And Bedrooms

Find the average number of bathrooms and bedrooms for each cityâ€™s property types. Output the result along with the city name and the property type.

```python
airbnb_search_details = read_csv('airbnb_search_details')
airbnb_search_details.head()
```

```python
result = airbnb_search_details.groupby(['city','property_type'])[['bedrooms','bathrooms']].mean().reset_index().rename(index=str, columns={"bedrooms": "n_bedrooms_avg", "bathrooms": "n_bathrooms_avg"})
result.head()
```
